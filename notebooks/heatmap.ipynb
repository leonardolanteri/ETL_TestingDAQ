```python
import sys
sys.path.append("..")
import matplotlib.pyplot as plt 
import numpy as np
import mplhep as hep
plt.style.use(hep.style.CMS)
import yaml
```

```python
def yaml_to_heatmap(path: str) -> None:
    with open(path) as f:
        data = yaml.load(f, yaml.Loader)

    plt.style.use(hep.style.CMS)

    fig, ax = plt.subplots(1,1,figsize=(12,12))
    cax = ax.matshow(data)
    ax.set_ylabel(r'$Row$')
    ax.set_xlabel(r'$Column$')
    fig.colorbar(cax,ax=ax)

    for (row, col), weight in np.ndenumerate(data):
        # x = col, y = row
        plt.text(col, row, f'{int(weight)}', ha='center', va='center', color='black' , fontsize="xx-small")


```
